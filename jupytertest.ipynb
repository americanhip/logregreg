```python
import linreg as lg
from pycaret.classification import ClassificationExperiment, add_metric
import pandas as pd

datafull = lg.df_procat("mHHS")

headers_keep = ['WC', 'Age at Sx', 'BMI', 'GM Repair', 'Tonnis Grade (Pre-op)', 'Ischial Spine (Pre-op)', 'Crossover (Pre-op)', 'Lateral CEA (Pre-op)', 'Acetabular Inclination (Pre-op)', 'Joint Space - Medial (Pre-op)', 'Joint Space - Central (Pre-op)', 'Joint Space - Lateral (Pre-op)', 'Neck-Shaft Angle (Pre-op)', 'Coxa Profunda (Pre-op)', 'Anterior CEA (Pre-op)', 'Alpha Angle (Pre-op)', 'Femoral Offset (Pre-op)', 'Lateral Imping', 'Side_R', 'Anterior Impinge_Negative', 'Anterior Impinge_Positive', 'Sex_Male', 'Pre mHHS', 'dmHHS']

data = datafull.loc[:, headers_keep]
#data = lg.impute(data)

print(list(data))
print("yay")
```

```python
s = ClassificationExperiment()
s.setup(data, target = 'dmHHS', session_id = 123, numeric_iterative_imputer= 'knn', categorical_imputation= 'drop')
# add metric
from sklearn.metrics import log_loss, accuracy_score
# s.add_metric('logloss', 'Log Loss', log_loss, greater_is_better = False)
s.add_metric('accuracy', 'Accuracy', accuracy_score, greater_is_better = True)
best = s.compare_models()
s.evaluate_model(best)
s.predict_model(best)
```
